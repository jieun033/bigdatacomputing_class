# 1. 넘파일 다차원 배열(ndarray) 객체 생성

- 리스트을 이용하여 다차원 배열 생성

# 9.  넘파이 다차원 배열의 슬라이싱을 이용한 논리인덱싱
- 특정한 조건을 만족하는 배열의 원소값만을 추출하여 1차원 배열 객체로 반환
- 브로드캐스팅하여 병렬 연산 처리


1. 주어진 배열을 이용하여 조건식을 만족하는 동일한 형상을 가지는 불리언 배열을 생성

2.   주어진 배열에 불리언 배열을 인덱스로 지정

- cf. np.where(조건식, x, y)
  - 주어진 배열의 구조는 유지하면서 조건식에 맞는 원소를 추출하는 기법
  - if - else 동작
  - x는 조건식이 True인 경우 처리
  - y는 조건식이 False인 경우 처리




```python
import numpy as np

arr = np.arange(100).reshape(5, -1)
print(arr)

# 짝수만 추출하여 1차원 배열로 생성
arr_bool = arr % 2 == 0
print("\n", arr_bool)

arr_even = arr[arr_bool]
print("\n", arr_even)

# 20보다 큰 요소들만 추출하여 1차원 배열로 생성
arr_greaterthan20 = arr[arr > 20]
print("\n", arr_greaterthan20)

# 20보다 큰 요소들만 추출
arr_filter = np.where(arr > 20, arr, 0) #true면 요소 그대로 출력, false면 0으로 마스킹됨
print("\n", arr_filter)
```

# 14. 다차원 배열 프로그래밍
- 벡터화 연산 (데이터병렬 처리)
  - 배열 간 연산
  - 리덕션
    - np.sum(arr, axis=)
    - np.mean(arr, axis=)
    - arr.sum(axis=0)
    - arr.sum(axis=1)
    - arr.sum(axis=(0,1))
  - 조건 리덕션 원리 (Boolean to Integer)
      - NumPy에서 리덕션이 강력한 이유는 **논리값(True/False)**을 연산하는 순간 **숫자(1/0)**로 자동 변환.

      - np.sum(condition) : 조건이 참(True)인 개수를 구함

      - np.mean(condition) : 조건이 참(True)인 비율을 구함 (예: 0.7이면 70%가 참)
      - 이 리덕션 기법들을 활용해서 실제 데이터셋에서 "이상치(Outlier)"를 찾기

```python
# 3인의 국영수 점수에 대한 데이터를 난수로 생성해서 원본 데이터, 과목 점수 평균, 학생 점수의 평균에 대한 세 종류의 그래프 분석
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import warnings # 경고 메시지 숨기기
import os

# 환경 설정 (경고 차단 및 한글 폰트)
warnings.filterwarnings('ignore')
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
if not os.path.exists(font_path):
  !sudo apt-get install -y fonts-nanum > /dev/null
font_prop = fm.FontProperties(fname=font_path)

np.random.seed(42)
scores = np.random.randint(60, 100, 12).reshape(3, -1)
subjects = ["국어", "영어", "수학", "역사"]
students = ["홍길동", "김영희", "김철수"]

subject_avg = np.mean(scores, axis = 0).round(2)
student_avg = np.mean(scores, axis = 1).round(2)

# 복수 개 그래프로 시각화
plt.figure(figsize = (16, 6))
plt.suptitle("원본 데이터 및 리덕션 데이터 분석", fontproperties = font_prop, fontsize = 20)

# (1) 원본 데이터 열지도
plt.subplot(1, 3, 1) # 그래프의 위치
im = plt.imshow(scores, cmap = "YlGnBu", aspect='auto')
plt.title("원본 성적 분포", fontproperties = font_prop, fontsize = 15)
plt.xlabel("과목명", fontproperties = font_prop)
plt.ylabel("학생명", fontproperties = font_prop)
plt.xticks(np.arange(len(subjects)), subjects, fontproperties = font_prop)
plt.yticks(np.arange(len(students)), students, fontproperties = font_prop)
cbar = plt.colorbar(im)
cbar.set_label("점수", fontproperties = font_prop)
# (1-1) 특정 좌표에 점수 텍스트를 표기: plt.text(x, y, str())
for i in range(3): # 행 (y축)
  for j in range(4): # 열 (x축)
    plt.text(j, i, scores[i, j], ha = "center", va = "center", color = "black", fontsize = 12)

# (2) 과목별 평균 시각화: 막대 그래프
plt.subplot(1, 3, 2)
plt.bar(subjects, subject_avg, color = ['red', 'green', 'blue', 'yellow'], edgecolor = 'black')
plt.title("과목별 평균 분포", fontproperties = font_prop)
plt.xlabel("과목명", fontproperties = font_prop)
plt.ylabel("평균 점수", fontproperties = font_prop)
plt.xticks(np.arange(len(subjects)), subjects, fontproperties = font_prop)
for i, score in enumerate(subject_avg): # 막대 상단에 평균 점수 표시
  plt.text(i, score, str(score), ha = "center", va = "bottom", color = "black", fontsize = 12)

# (3) 학생별 평균 시각화: 수평막대 그래프
plt.subplot(1, 3, 3)
plt.barh(students, student_avg, color = ['lightgreen'], edgecolor = 'black')
plt.title("학생별 평균 분포", fontproperties = font_prop)
plt.xlabel("평균 점수", fontproperties = font_prop)
plt.ylabel("학생명", fontproperties = font_prop)
plt.yticks(fontproperties = font_prop)
for i, score in enumerate(student_avg): # 수평막대 우측에 평균 점수 표시
  plt.text(score + 1, i, str(score), va = "center", color = "black", fontsize = 12, fontweight = 'bold')

plt.tight_layout()
plt.show()
```

# 15. 넘파일 배열의 결합
- np.concatenate(튜플)
- np.vstack(튜플 또는 리스트)
- np.hstack(튜플 또는 리스트)
- np.column_stack(리스트)


```python
# 1. 1차원 배열 간 결합
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# 세로로 쌓기 (수직 방향) -> 행 추가해서 2차원 배열
res_stack = np.vstack((a, b)) # ((반드시 튜플로 쓸 것))
print(res_stack)

# 가로로 붙이기 (수평 방향) -> 열 추가해서 1차원 배열
res_hstack = np.hstack((a, b))
print(res_hstack)
res_concat = np.concatenate((a, b), axis=0) # axis=0은 생략 가능
print(res_concat)

# 주의: column_stack() -> 2차원 배열
res_col = np.column_stack((a, b))
print(res_col)

# ------------------------------------------------------------------------
# 2. 형태가 다른 배열 간 결합
arr_1 = np.arange(1, 4) # (3, )
arr_2 = np.arange(4, 10).reshape(2, 3)
print("============\n", arr_1)

# 방법 1
# (1) arr_1 -> (1, 3) : 열의 개수를 일치
arr_1_2d = arr_1.reshape(1, 3)
print("============\n", arr_1_2d)
combined_2d = np.concatenate((arr_1_2d, arr_2), axis=0)
print("============\n", combined_2d)

# 방법 2
v_combined = np.vstack((arr_1, arr_2))
print("============\n", v_combined)
```
